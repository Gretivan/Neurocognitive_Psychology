Oh oh, that doesn't look good...they're not spatially aligned and thus we can't really extract data from the `thickness image` based on the `atlas`. We need to change things up a bit...this entails:

- get a different version of the `thickness data`
- get the `transformations` we need to transform the `thickness data` from a given `participant`'s `space` to a/the `template space`, specifically `MNI152NLin2009cAsym`

### Get different version of cortical thickness data

```python
url = "https://fcp-indi.s3.amazonaws.com/data/Projects/ABIDE/Outputs/mindboggle_swf/mindboggle/ants_subjects/sub-0050003/antsCorticalThickness.nii.gz" 
urllib.request.urlretrieve(url, '/Users/peerherholz/Desktop/sub-0050003_desc-corticalthickness.nii.gz')

```

### Get transformations from participant to template space

```python
url = "https://fcp-indi.s3.amazonaws.com/data/Projects/ABIDE/Outputs/fmriprep/fmriprep/sub-0050003/anat/sub-0050003_from-T1w_to-MNI152NLin2009cAsym_mode-image_xfm.h5" 
urllib.request.urlretrieve(url, '/Users/peerherholz/Desktop/sub-0050003_from-T1w_to-MNI152NLin2009cAsym_mode-image_xfm.h5')
```

### Get the template image

```python


mni152 = tflow.get('MNI152NLin2009cAsym', desc=None, resolution=1,
                    suffix='T1w', extension='nii.gz')
mni152
```

### Apply the transformations to the participant cortical thickness image, transforming it to template space

For this you need to install [ANTs](https://github.com/ANTsX/ANTs/releases), try the first way outlined [here](https://miykael.github.io/nipype-beginner-s-guide/installation.html#ants).

```python
%%bash
antsApplyTransforms -i /Users/peerherholz/Desktop/sub-0050003_desc-corticalthickness.nii.gz -r /Users/peerherholz/.cache/templateflow/tpl-MNI152NLin2009cAsym/tpl-MNI152NLin2009cAsym_res-01_T1w.nii.gz -t /Users/peerherholz/Desktop/sub-0050003_from-T1w_to-MNI152NLin2009cAsym_mode-image_xfm.h5 -o /Users/peerherholz/Desktop/sub-0050003_space-MNI152NLin2009cAsym_desc-corticalthickness.nii.gz
```

### Check alignment by plotting cortical thickness data and overlaying Destrieux atlas

```python
fig = plot_img('/Users/peerherholz/Desktop/sub-0050003_space-MNI152NLin2009cAsym_desc-corticalthickness.nii.gz', cmap='magma')
fig.add_contours(destrieux_atlas, cmap='Greys')
```

Looks ok, let's check if we can now extract things accordingly.

### Try the masker with the new data

```python
destrieux_masker = NiftiLabelsMasker(labels_img=destrieux_atlas)
thickness_destrieux = destrieux_masker.fit_transform('/Users/peerherholz/Desktop/sub-0050003_space-MNI152NLin2009cAsym_desc-corticalthickness.nii.gz')
```

```python
thickness_destrieux.shape
```

```python
len(np.unique(destrieux_atlas.dataobj))
```

Please note that one label is `background` and thus things match up!

### How to proceed?

The above steps need to be repeated for each `participant`, ie:

- get the new thickness data
- get the spatial transforms
- apply the spatial transforms

NB: I think the `participant ID`s are identical, e.g. here `0050003`, and you just need to exchange the site `identifier` to `sub`, e.g. here `Pitt_0050003` -> `sub-0050003`.
