### Data and Project Management 

The purpose of this notebook is to download, assess, and explore the data that will be worked with in this Neurocognitive psychology project. The dataset used is the Autism Brain Imaging Data Exchange II (ABIDE II) dataset. ABIDE II was developed as a continuation to the ABIDE I dataset as part of the 1000 Functional Connectomes Project, to further encourage discovery science on the brain connectome in Autism Spectrum Disorder. This dataset consists of MRI data of patients with and without Autism, from various different sites around the world. From this dataset we will be using the anatomical data, and the CSV file with all phenotypic information for the participants.
